```python
import numpy as np
import scipy.integrate as inte 
#import matplotlib.ticker as mticker

#2104.05013
#va = sorate dar emtedad a
#vb = sorate dar emtedad b
#Bz = meydan magnetic dar emtedad z
#Ba = meydan dar emtedad a
#Bb = medyan dar emtedad b
#x = t/t_ew
#yR = assymetry lepton right
#yB = assymetry baryoun (yani hame quarks)
def S2(x,u):
    Bz,Ba,Bb,yR,yB,yL,va,vb=u
    #table 1
    F0 = 25.78
    F1 = 77.79
    F2 = 5.6 * 1e-6
    F3 = 356 * 1e+6
    F4 = 77809 * 1e+8
    F5 = 0.304                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                              
    F6 = 7.12
    aY = 0.0095 #alpha_y finite structure constant (zire table 1)
    Gam0 = 121 #rate vakonesh \Gamma_0
    k2=1 #k'' zire table 1 ke k=10**-7
    Pi=np.pi

    dyR = F0*((Ba/1e+20)**2 + (Bb/1e+20)**2)* x**(3/2)\
        - F1*(yR - yL/2 + (3/8)*yB)*((Bz/1e+20)**2 + (Ba/1e+20)**2 +\
            (Bb/1e+20)**2) * x**(3/2) -\
                F2*(yR**2 - yL**2)*(va*Ba/1e+20 + vb*Bb/1e+20) * x**(1/2)\
                        - Gam0 * (1 - x) * (yR - yL)/x**(1/2)

    dyL = -F0/4*((Ba/1e+20)**2 + (Bb/1e+20)**2) * x**(3/2)\
        + F1/4*(yR - yL/2 + (3/8)*yB)*((Bz/1e+20)**2 + (Ba/1e+20)**2\
            + (Bb/1e+20)**2) * x**(3/2) +\
                F2/4*(yR**2 - yL**2)*(va*Ba/1e+20 + vb*Bb/1e+20) * x**(1/2)\
                        + Gam0*(1 - x)*(yR - yL) /(2 * x**(1/2))

    dyB = (3/2)*F0*((Ba/1e+20)**2 +\
        (Bb/1e+20)**2) * x**(3/2) - \
            (3/2)*F1*(yR - yL/2 + (3/8)*yB)*((Bz/1e+20)**2\
                + (Ba/1e+20)**2 + (Bb/1e+20)**2) * x**(3/2)\
                    - (3/2)*F2*(yR**2 - yL**2)*(va*Ba/1e+20 + vb*Bb/1e+20) * x**(1/2)

    dBz = -Bz/x

    dBa = 356*k2*(aY*(yR - yL/2 + (3/8)*yB)/Pi\
        - k2/1e+3) * Ba/x**(1/2) + F3*vb*Bz/x**(1/2) \
            + F4*(yR**2 - yL**2)*va/x**(3/2) - Ba/x

    dBb = 356*k2*(aY*(yR - yL/2 + (3/8)*yB)/Pi - 1/1e+3)*Bb/x**(1/2) \
        - F3*va*Bz/x**(1/2) + F4*(yR**2 - yL**2) * vb/x**(3/2) - Bb/x

    dva = F5*(Bz/1e+20)*(Bb/1e+20) * x**(3/2) - F6*va/(aY**2 * x**(1/2))

    dvb = -F5*(Bz/1e+20)*(Ba/1e+20) * x**(3/2) - F6*vb/(aY**2 * x**(1/2))
    return np.array([dBz,dBa,dBb,dyR,dyB,dyL,dva,dvb])

Bz0,Ba0,Bb0,yR0,yB0,yL0,va0,vb0=1e+23,0,0,0,0,0,1e-2,1e-2
S20=[Bz0,Ba0,Bb0,yR0,yB0,yL0,va0,vb0]

s= inte.solve_ivp(fun=S2,t_span=(1e-4,1),y0=S20,method='LSODA') #t haman  hast x=t/t_ew

Bz,Ba,Bb,yR,yB,yL,va,vb=s.y
x=s.t

nR = 3.55929 * 1e-7 * yR
nL = 3.55929 * 1e-7 * yL
nB = 3.55929 * 1e-7 * yB

By = (Ba**2 + Bb**2 + Bz**2)**0.5

s
```

```python
import matplotlib.pyplot as plt

def plot(x,y,title):
  fig, ax1 = plt.subplots(figsize=(10,5))
  ax1.plot(x,y)
  ax1.grid(True)
  plt.yscale('symlog')
  plt.xscale('log')
  plt.xlabel('x')
  plt.ylabel(title)
  plt.title('Diagram of ' + title + ' per $x = t/t_{ew}$')

  left, bottom, width, height = [0.5, 0.6, 0.2, 0.2]
  ax2 = fig.add_axes([left, bottom, width, height])
  ax2.plot(x[50:100],y[50:100])
  plt.xlabel('x')
  plt.ylabel(title)
  plt.yscale('symlog')
  plt.xscale('log')

  plt.show()
```

```python
plot(x,va,'$v_a$')
plot(x,vb,'$v_b$')
```

```python
plot(x,Ba,'$B_a$')
plot(x,Bb,'$B_b$')
plot(x,By,'$B_y$')
```

```python
plot(x,nB,'$n_B$')
plot(x,nL,'$n_L$')
plot(x,nR,'$n_R$')
```
